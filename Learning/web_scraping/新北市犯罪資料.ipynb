```
"""
檔名：新北市犯罪資料.py
功能：網頁資料擷取練習
學習重點：requests、BeautifulSoup、urllib3 網頁爬取與解析
"""
import requests as rq
import pandas as pd
from xml.etree import ElementTree as ET
```

```
# 犯罪資料
url = "https://data.ntpc.gov.tw/api/datasets/8A32C6B5-46FC-4FAC-B3A4-317B9998BFD7/xml/preview"
```

```
r = rq.request("GET", url)
```

```
# 將得到的資料轉換成一棵元素樹（ElementTree）
xml_data = ET.fromstring(r.content)
```

```
list_data = []  # 用以存放全部資料的資料集串列
```

```
for record in xml_data.iter("row"):  # 針對每一筆資料
    lst = []
    for col in record.iter():
        lst.append(col.text)
    # lst[0]=None, lst[1]=案類, lst[2]=年份, lst[3]=發生日期, lst[4]=發生地區
    list_data.append(lst[1:5])
```

```
# 將資料集串列轉換成 DataFrame
df = pd.DataFrame(list_data, columns=["案類", "年份", "發生日期", "發生地區"])
df.index += 1  # 資料順序從1開始
```
