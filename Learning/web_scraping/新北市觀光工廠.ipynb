```
"""
檔名：新北市觀光工廠.py
功能：網頁資料擷取練習
學習重點：requests、BeautifulSoup、urllib3 網頁爬取與解析
"""
import requests as rq
from xml.etree import ElementTree
import pandas as pd
```

```
# 新北市觀光工廠
url = "https://data.ntpc.gov.tw/api/datasets/57EB9B00-979C-44BB-A4EE-CC55BDF1488A/xml/preview"
```

```
r = rq.request("GET", url)
```

```
# 將資料的內容轉換成一棵元素樹（ElementTree）
tree = ElementTree.fromstring(r.content)
```

```
list_data = []  # 用以存放轉換成串列的資料
```

```
for i in tree.iter("row"):  # 針對 ElementTree 的每一個 row 標籤
    single_record = []  # 儲存一筆轉換成串列的資料
    for j in i.iter():
        if (
            j.tag == "organization"
            or j.tag == "introduction"
            or j.tag == "localcallservice"
            or j.tag == "adress"
        ):
            single_record.append(j.text)
    list_data.append(single_record)
```

```
# 將資料轉換成 DataFrame
df = pd.DataFrame(list_data, columns=["名稱", "特色", "電話", "地址"])
df.index += 1  # 資料順序從1開始
```
