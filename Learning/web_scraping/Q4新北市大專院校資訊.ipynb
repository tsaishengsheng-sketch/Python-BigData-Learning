```
"""
檔名：Q4新北市大專院校資訊.py
功能：網頁資料擷取練習
學習重點：requests、BeautifulSoup、urllib3 網頁爬取與解析
"""
import requests
```

```
# 新北市重要地表資訊－大專院校
url = "https://data.ntpc.gov.tw/api/datasets/6DCFF24A-838C-40FB-A9DF-F1160AFAFE84/json/preview"
```

```
r = requests.get(url, timeout=10)
json_data = r.json()
```

```
print(f"回傳資料筆數：{len(json_data)}\n")
print(f"{'名稱':<20} {'地址':<35} {'聯絡電話':<15} {'網站'}")
print("-" * 90)
```

```
for item in json_data:
    name = item.get("地標名稱", "")
    address = item.get("地址", "")
    tel = item.get("電話", "")
    website = item.get("網址") or ""
    update = item.get("更新日期", "")
    print(f"{name:<20} {address:<35} {tel:<15} {website}")
```
