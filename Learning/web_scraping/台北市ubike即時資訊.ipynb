```
"""
檔名：台北市ubike即時資訊.py
功能：網頁資料擷取練習
學習重點：requests、BeautifulSoup、urllib3 網頁爬取與解析
"""
import sys
import json
import requests
```

```
# 台北市 YouBike 2.0 即時資訊
url = "https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json"
```

```
data = requests.get(url).content.decode("utf-8")
jsondata = json.loads(data)
```

```
for i in range(len(jsondata)):
    detail = jsondata[i]
    sna = detail["sna"]  # 場站名稱（中文）
    sbi = detail["available_rent_bikes"]  # 場站目前可借車輛數量
    bemp = detail["available_return_bikes"]  # 空位數量
    act = detail["act"]  # 全站禁用狀態
    mday = detail["mday"]  # 資料更新時間

    row = (
        "站名:"
        + sna
        + ","
        + "可借:"
        + str(sbi)
        + ","
        + "可停:"
        + str(bemp)
        + ","
        + "場站運作中:"
        + act
    )

    print(row.encode(sys.stdin.encoding, "replace").decode(sys.stdin.encoding))
```
