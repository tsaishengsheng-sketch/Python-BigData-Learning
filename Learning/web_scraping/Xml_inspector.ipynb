```
"""
檔名：Xml_inspector.py
功能：網頁資料擷取練習
學習重點：requests、BeautifulSoup、urllib3 網頁爬取與解析
"""
import requests as rq
from xml.etree import ElementTree
```

```
# ── 設定區（每次使用只需修改這裡）──────────────────────
URL = "https://your-api-url-here"  # 換成目標網址
TARGET_TAG = "row"  # 換成要查看的父標籤（不確定就留 'row'）
PREVIEW_COUNT = 3  # 預覽幾筆資料
# ────────────────────────────────────────────────────────
```

```
r = rq.request("GET", URL)
tree = ElementTree.fromstring(r.content)
```

```
print(f"根標籤：{tree.tag}\n")
```

```
# 模式一：指定父標籤，查看其下所有子標籤
count = 0
for i in tree.iter(TARGET_TAG):
    if count >= PREVIEW_COUNT:
        break
    print(f"--- {TARGET_TAG} {count + 1} ---")
    for j in i.iter():
        print(f"  tag: {j.tag:<25} text: {j.text}")
    count += 1
```

```
# 如果沒找到任何資料，改用全樹掃描
if count == 0:
    print(f"找不到標籤 '{TARGET_TAG}'，改為列出整棵樹前50個標籤：\n")
    for i, j in enumerate(tree.iter()):
        if i >= 50:
            break
        print(f"  tag: {j.tag:<25} text: {j.text}")
```
