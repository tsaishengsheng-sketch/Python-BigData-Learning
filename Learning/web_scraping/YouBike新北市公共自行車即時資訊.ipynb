```
"""
檔名：YouBike新北市公共自行車即時資訊.py
功能：網頁資料擷取練習
學習重點：requests、BeautifulSoup、urllib3 網頁爬取與解析
"""
import requests as rq
```

```
# 開放資料：YouBike 2.0 即時資訊
url = "https://apis.youbike.com.tw/json/station-yb2.json"
```

```
html_content = rq.get(url)
json_data = html_content.json()
```

```
# 只顯示前10筆
for item_detail in json_data[:10]:
    print_info = (
        "站點："
        + item_detail["name_tw"]
        + "，"
        + "地址："
        + item_detail["address_tw"]
        + "，"
        + "總停車格："
        + str(item_detail["parking_spaces"])
        + "，"
        + "場站目前車輛數量："
        + str(item_detail["available_spaces"])
        + "，"
        + "空位數量："
        + str(item_detail["empty_spaces"])
        + "，"
        + "資料更新時間："
        + item_detail["updated_at"]
    )
    print(print_info)
```
