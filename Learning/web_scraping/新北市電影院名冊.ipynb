```
"""
檔名：新北市電影院名冊.py
功能：網頁資料擷取練習
學習重點：requests、BeautifulSoup、urllib3 網頁爬取與解析
"""
import requests as rq
import csv
import pandas as pd
```

```
# 新北市電影院名冊
url = "https://data.ntpc.gov.tw/api/datasets/61C99F42-8A90-4ADC-9C40-BA9E0EA097AA/csv/preview"
```

```
r = rq.request("GET", url)
```

```
# 將 CSV 格式的字串轉換成二維串列
data = list(csv.reader(r.text.splitlines(), delimiter=",", quotechar='"'))
```

```
# 將資料集串列轉換成 DataFrame
df = pd.DataFrame(data[1 : len(data) - 1], columns=["名稱", "地址", "電話號碼", "廳數"])
df.index += 1  # 資料順序從1開始
```
