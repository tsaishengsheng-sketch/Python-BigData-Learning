```
"""
檔名：台灣證券交易所日成交量前二十名.py
功能：網頁資料擷取練習
學習重點：requests、BeautifulSoup、urllib3 網頁爬取與解析
"""
import pandas as pd
import requests as rq
```

```
# 日成交量前二十名證券
url = "https://www.twse.com.tw/exchangeReport/MI_INDEX20?response=json&date=&_=1513501982216"
```

```
r = rq.get(url)
```

```
# 將請求結果轉換成 json 格式
json_data = r.json()
```

```
# 將 json 資料轉換成 pandas DataFrame
df = pd.DataFrame(json_data["data"], columns=json_data["fields"])
print(df)
```

```
# 顯示 '漲跌(+/-)' 欄位
print(df["漲跌(+/-)"])
```

```
# 處理 '漲跌(+/-)' 欄位：去掉 '+' 或 '-' 以外的字元
for i in range(len(df["漲跌(+/-)"])):
    if df.loc[i, "漲跌(+/-)"] == "":
        continue
    else:
        index_of_target = df.loc[i, "漲跌(+/-)"].index(">") + 1
        target = df.loc[i, "漲跌(+/-)"][index_of_target]
        df.loc[i, "漲跌(+/-)"] = target  # 使用 df.loc 避免警告
```

```
print(df["漲跌(+/-)"])
```
