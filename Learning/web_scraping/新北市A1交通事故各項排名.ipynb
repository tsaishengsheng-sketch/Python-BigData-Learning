```
"""
檔名：新北市A1交通事故各項排名.py
功能：網頁資料擷取練習
學習重點：requests、BeautifulSoup、urllib3 網頁爬取與解析
"""
import json
import requests
import pandas as pd
```

```
# 設定顯示前 n 名
nShowTop = 5
```

```
def DisplayTopN(Title, Captions, Data):
    print("<<<< ", Title, "前", nShowTop, "名 >>>>>")
    _nLoop = 1
    for _i, _row in Data.iterrows():
        print("第", _nLoop, "名 -", Captions[_row.name], " ", _row[0], "次")
        _nLoop = _nLoop + 1
        if _nLoop > nShowTop:
            break
    print()
```

```
def BuildColumnList(Captions):
    _cols = []
    for _i in range(1, len(Captions) - 1):
        _cols.append("quantity{0}".format(_i))
    return _cols
```

```
def LoadJsonAndSort(url, cols):
    _data = requests.get(url).content.decode("utf-8")
    _jsondata = json.loads(_data)
    _df = pd.DataFrame(_jsondata, columns=cols)
    _df = _df.replace([""], [0]).fillna(0).astype(int).sum()
    _sorted = _df.sort_values(ascending=False).to_frame()
    return _sorted
```

```
# 每月新北市 A1 交通事故車輛種類
url = "https://data.ntpc.gov.tw/api/datasets/03A46914-B774-4C1C-BF75-E6EECE1E311C/json"
captions = {
    "quantity1": "營業大客車",
    "quantity2": "自用大客車",
    "quantity3": "營業小客車",
    "quantity4": "自用小客車",
    "quantity5": "營業大貨車",
    "quantity6": "自用大貨車",
    "quantity7": "自用小貨車",
    "quantity8": "重型機車",
    "quantity9": "輕型機車",
    "quantity10": "軍用小客車",
    "quantity11": "軍用吉普車",
    "quantity12": "軍用大貨車",
    "quantity13": "其他",
}
sorted_data = LoadJsonAndSort(url, BuildColumnList(captions))
DisplayTopN("每月新北市A1交通事故車輛種類", captions, sorted_data)
```
