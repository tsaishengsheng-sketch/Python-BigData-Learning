```
"""
檔名：verify_matrix_even_ones.py
功能：多維串列應用 - Verify Matrix Even Ones
學習重點：矩陣運算、巢狀迴圈與表格處理
"""
import random
```

```
lst66 = []
for i in range(6):
    lst66.append([])
    for j in range(6):
        randNum = random.randint(0, 1)
        lst66[i].append(randNum)
```

```
for row in lst66:
    for value in row:
        print(f"{value:3d}", end=' ')
    print()
```

```
print()
for row in range(len(lst66)):
    count = sum(lst66[row])
    if count != 0:
        if count % 2 == 0:
            print(f"{row} row has even 1")
```

```
print()
for col in range(len(lst66[0])):
    total = 0
    for row in range(len(lst66)):
        total += lst66[row][col]
    if total % 2 == 0:
        if total != 0:
            print(f"{col} column has even 1")
```
