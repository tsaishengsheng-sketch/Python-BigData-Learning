```
for i in range(len(scores3Dim)):
    print(f"{(i+1):3d}", end='')
    for j in range(len(scores3Dim[i])):
        total = 0
        for k in range(len(scores3Dim[i][j])):
            total += scores3Dim[i][j][k]
        average = total / 3
        print(f"{average:9.2f}", end='')
    print()
```
