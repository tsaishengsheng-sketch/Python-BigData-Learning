```
"""
檔名：double_even_numbers.py
功能：多維串列應用 - Double Even Numbers
學習重點：矩陣運算、巢狀迴圈與表格處理
"""
import random
```

```
def evenDouble(lst2):
    for i in range(len(lst2)):
        for j in range(len(lst2[i])):
            if lst2[i][j] % 2 == 0:
                lst2[i][j] *= 2
    return lst2
```

```
def display(lst3):
    for row in lst3:
        for value in row:
            print(f"{value:5d}", end=' ')
        print()
```

```
def createlist():
    lst = []
    rows = eval(input('How many rows: '))
    columns = eval(input('How many columns: '))
    for i in range(rows):
        lst.append([])
        for j in range(columns):
            num = random.randint(1, 100)
            lst[i].append(num)
    return lst
```

```
def main():
    lst2 = createlist()
    print(f"Original list: ")
    display(lst2)
    print(f"After double even number")
    lst3 = evenDouble(lst2)
    display(lst3)
```

```
main()
```
