```
"""
檔名：simulate_grading_system.py
功能：多維串列應用 - Simulate Grading System
學習重點：矩陣運算、巢狀迴圈與表格處理
"""
import random
```

```
students = eval(input('How many students? '))
questions = eval(input('How many questions? '))
```

```
# 產生學生答案 (A-E)
studentAns = []
for i in range(students):
    studentAns.append([])
    for j in range(questions):
        # 產生 65-69 的整數 (A-E)
        ans = random.randint(65, 69)
        # 直接轉成字元存入
        studentAns[i].append(chr(ans))
        # 修正點：既然已經是字串，格式化改用 :3s 或直接不加 c
        print(f"{studentAns[i][j]:3s}", end=' ')
    print()
```

```
# 產生標準答案 (A-E)
standAns = []
print(f"correct answer: ")
for i in range(questions):
    n = random.randint(65, 69)
    standAns.append(chr(n))
    # 修正點：同上，改用 :3s
    print(f"{standAns[i]:3s}", end=' ')
print()
```

```
# 批改邏輯
for i in range(len(studentAns)):
    correctNum = 0
    for j in range(len(studentAns[i])):
        if studentAns[i][j] == standAns[j]:
            correctNum += 1
    print(f"Student #{i+1} score: {correctNum}")
```
