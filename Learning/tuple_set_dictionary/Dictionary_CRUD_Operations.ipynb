```
"""
檔名：Dictionary_CRUD_Operations.py
功能：進階資料結構練習
學習重點：不可變與可變資料類型、集合運算
"""
import random
```

```
def menu():
    # 使用 \n 縮減程式碼行數，達成單行輸出選單
    print('\nDictionary menu:\n1. insert\n2. delete\n3. display\n4. exit')
    n = eval(input('Enter your choice: '))
    return n
```

```
def add(d2):
    # 取得 Key 與 Value 並新增至字典，若 Key 重複則會直接覆蓋 (Update)
    key = eval(input('Enter a key: '))
    value = input('Enter a value: ')
    print(f"Adding... {key}: {value}")
    d2[key] = value
```

```
def delete(d2):
    key = eval(input('Enter a key: '))
    # 使用 in 關鍵字檢查 Key 是否存在，避免直接刪除導致程式崩潰
    if key in d2:
        # 使用 pop() 移除指定鍵值並回傳其值
        removed_value = d2.pop(key)
        print(f"Success: {key}: {removed_value} has been deleted.")
    else:
        # 當 Key 不存在時的錯誤處理邏輯
        print('Error: Key not found.')
```

```
def display(d2):
    # 印出當前字典的所有鍵值對狀態
    print(f"Current Dictionary Content: {d2}")
```

```
def main():
    # 宣告空字典 d 作為資料儲存容器
    d = {}
    while True:
        choice = menu()
        if choice == 1:
            add(d) # 新增或修改資料
        elif choice == 2:
            delete(d) # 刪除資料
        elif choice == 3:
            display(d) # 查閱資料
        elif choice == 4:
            print("System Exited.")
            break # 退出迴圈
        else:
            print('Invalid choice, please try again.')
```

```
if __name__ == "__main__":
    main()
```
