```
"""
檔名：Most_Frequent_Number_Finder.py
功能：進階資料結構練習
學習重點：不可變與可變資料類型、集合運算
"""
def main():
    # 取得使用者輸入字串並移除前後空格
    s = input("Enter the numbers: ").strip()
    
    # 使用串列生成式 (List Comprehension) 將分割後的字串轉為數字列表
    numbers = [eval (x) for x in s.split()]
    
    # 建立一個空字典用於存放數字與其出現次數
    dictionary = {} 
    
    # 遍歷數字列表進行計數
    for number in numbers:
        if number in dictionary:
            # 若數字已在字典中，計數加 1
            dictionary[number] += 1
        else:
            # 若是第一次出現，初始化計數為 1
            dictionary[number] = 1
            
    # 找出字典中所有「值」(出現次數) 的最大值
    maxCount = max(dictionary.values())

    # 將字典的鍵值對轉換為列表形式 (元組清單)
    pairs = list(dictionary.items())
    print(pairs)

    print("The numbers with the most occurrence are ", end="")
    
    # 遍歷元組清單，x 為數字，y 為次數
    for (x,y) in pairs:
        # 篩選出次數等於最大值的數字
        if y == maxCount:
            print(x, end=" ")
```

```
# 執行主程式
if __name__ == "__main__":
    main()
```
