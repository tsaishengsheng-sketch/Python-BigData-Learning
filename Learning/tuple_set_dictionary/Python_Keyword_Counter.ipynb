```
"""
檔名：Python_Keyword_Counter.py
功能：進階資料結構練習
學習重點：不可變與可變資料類型、集合運算
"""
import os
```

```
def main():
    # 1. 定義關鍵字集合 (Set) 
    # 使用集合 {...} 而非串列 [...] 是因為集合搜尋的時間複雜度為 O(1)，在大數據下極快
    keywords = {
        'and', 'del', 'or', 'not', 'while', 'for', 'with', 'break', 
        'True', 'False', 'elif', 'else', 'if', 'except', 'import', 
        'print', 'class', 'in', 'continue', 'finally', 'is', 'return', 
        'def', 'try'
    }

    # 2. 檔案路徑檢核循環 (延伸自上一章的 IOError 處理邏輯)
    while True:
        f1 = input('Enter a Python source code filename: ').strip()
        # 檢查檔案是否真的存在，避免 FileNotFoundError
        if os.path.exists(f1):
            break
        print(f"❌ 找不到檔案 '{f1}'，請確認路徑是否正確")

    # 3. 讀取與邏輯運算
    try:
        infile = open(f1, 'r')
        # split() 會自動以空白、換行切分字串成清單
        text = infile.read().split()
        
        # 初始化計數器 (這是你之前漏掉導致報錯的關鍵)
        count = 0  
        for word in text:
            # 使用 'in' 運算子在 Set 中搜尋關鍵字
            if word in keywords:
                count += 1
        
        # 4. 格式化輸出結果
        print(f"✅ There are {count} keywords in {f1}")
        
        # 關閉檔案串流
        infile.close()
        
    except Exception as e:
        print(f"發生非預期錯誤: {e}")
```

```
if __name__ == "__main__":
    main()
```
