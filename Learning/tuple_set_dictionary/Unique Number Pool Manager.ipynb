```
"""
檔名：Unique Number Pool Manager.py
功能：進階資料結構練習
學習重點：不可變與可變資料類型、集合運算
"""
import random
```

```
def menu():
    print()
    print('Set menu:')
    print('1. insert')
    print('2. delete')
    print('3. display')
    print('4. exit')
    # 接收使用者輸入並轉換為數值
    n = eval(input('Enter your choice: '))
    return n 
```

```
def add(s2):
    # 隨機產生 1 到 49 之間的整數
    x = random.randint(1, 49)
    print(f"generating...{x}")
    # 利用集合 add 方法存入，若重複則自動忽略
    s2.add(x)
```

```
def delete(s2):
    # 詢問使用者要刪除的特定號碼
    x = eval(input('Enter a number: '))
    # 成員檢查：確保號碼在集合內才執行刪除，避免拋出 KeyError 錯誤
    if x in s2:
        s2.remove(x)
    else:
        # 若號碼不在集合中，印出提示訊息
        print('There is not invalid number.')
```

```
def display(s2):
    # 直接印出當前集合內容
    print(s2)
```

```
def main():
    # 建立一個初始的空集合
    s = set()
    while True:
        # 呼叫選單並取得回傳的選項編號
        choice = menu()

        if choice == 1:
            add(s) # 執行隨機新增
        elif choice == 2:
            delete(s) # 執行手動刪除
        elif choice == 3:
            display(s) # 執行顯示集合
        elif choice == 4:
            # 跳出迴圈，結束程式
            break
        else:
            # 處理 1-4 以外的無效輸入
            print('Invalid choice')
```

```
# 執行主程式
main()
```
