```
"""
檔名：Python_Source_Keyword_Frequency_Inversion.py
功能：進階資料結構練習
學習重點：不可變與可變資料類型、集合運算
"""
def main():
    # 定義 Python 關鍵字集合，用於快速比對
    keywords = {'and', 'del', 'or', 'not', 'while',
                'for', 'with', 'break', 'True', 'False',
                'elif', 'else', 'if', 'break', 'except',
                'import', 'print', 'class', 'in',
                'continue', 'finally', 'is', 'return',
                'def', 'try'}

    # 取得原始碼檔案名稱
    f1 = input('Enter a Python source code filename: ').strip()

    # 以讀取模式開啟檔案
    infile = open(f1, 'r')

    # 讀取全文並根據空白字元切分為單字列表
    text = infile.read().split()

    dictionary = {}
    # 遍歷文本中的每個單字
    for word in text:
        # 若單字屬於 Python 關鍵字集合
        if word in keywords:
            # 統計頻率：若已在字典中則累加，否則初始化為 1
            if word in dictionary:
                dictionary[word] += 1
            else:
                dictionary[word] = 1

    # 輸出原始的關鍵字統計字典
    print(dictionary)

    # 將字典轉換為 (key, value) 元組列表
    pairs = list(dictionary.items())
    
    # 使用串列生成式將 [word, count] 倒置為 [count, word]
    # 這步通常是為了方便後續使用 sort() 依照「出現次數」進行排序
    newPairs = [[y, x] for (x, y) in pairs] 
    
    # 輸出倒置後的列表
    print(newPairs)
```

```
# 執行主程序
if __name__ == "__main__":
    main()
```
