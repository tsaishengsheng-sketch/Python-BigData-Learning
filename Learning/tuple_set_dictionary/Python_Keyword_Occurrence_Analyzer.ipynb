```
"""
檔名：Python_Keyword_Occurrence_Analyzer.py
功能：進階資料結構練習
學習重點：不可變與可變資料類型、集合運算
"""
def main():
    # 定義要統計的 Python 關鍵字集合 (Set)
    keyword = {'and', 'del', 'or', 'not', 'while', 'for', 'with', 'break', 'True', 'False', 'else', 'if', 'except', 'import', 'print', 'class', 'in', 'continue', 'finally', 'is', 'return', 'def', 'try'}

    # 取得使用者輸入的檔名並移除前後空格
    f1 = input('Enter a Python source code filename: ').strip()

    try:
        # 開啟檔案進行讀取
        infile = open(f1, 'r')
        # 讀取全文並以空白字元切割成單字列表
        text = infile.read().split()

        dictionary = {}
        for word in text:
            # 僅針對屬於關鍵字集合的單字進行統計
            if word in keyword:
                # 修正：使用 get 處理第一次出現的情況，避免 KeyError
                dictionary[word] = dictionary.get(word, 0) + 1
        
        # 關閉檔案
        infile.close()

        # 輸出最終統計字典
        print("\nKeyword Frequency Result:")
        print(dictionary)

    except FileNotFoundError:
        print("Error: File not found. Please check the filename.")
```

```
if __name__ == "__main__":
    main()
```
