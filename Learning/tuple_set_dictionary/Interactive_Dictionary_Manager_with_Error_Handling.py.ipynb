```
"""
檔名：Interactive_Dictionary_Manager_with_Error_Handling.py.py
功能：進階資料結構練習
學習重點：不可變與可變資料類型、集合運算
"""
import random
```

```
def menu():
    """顯示選單並安全地取得使用者選項"""
    print('\nDictionary menu:\n1. insert\n2. delete\n3. modify\n4. display\n5. exit')
    try:
        # 將 eval 改為 int 以避免安全性風險與解析錯誤
        n = int(input('Enter your choice: '))
        return n
    except ValueError:
        # 當輸入非數字時，回傳 0 以觸發 else 的 Invalid choice
        return 0
```

```
def add(d2):
    """新增鍵值對，並確保 Key 為整數型態"""
    try:
        key = int(input('Enter a key: '))
        value = input('Enter a value (String): ')
        # 使用 f-string 格式化輸出字典內容
        print(f'generating ... {{{key}: {value}}}')
        d2[key] = value
    except ValueError:
        print("Invalid input, key must be an integer.")
```

```
def delete(d2):
    """根據鍵名刪除資料，包含存在性檢查"""
    try:
        key = int(input('Enter a key: '))
        # 成員檢查：避免刪除不存在的 Key 導致 KeyError
        if key in d2:
            del d2[key]
            print(f"Key {key} deleted.")
        else:
            print(f'Key {key} not found.') 
    except ValueError:
        print("Invalid input.")
```

```
def modify(d2):
    """更新已存在的鍵值內容"""
    try:
        key = int(input('Enter a key: '))
        if key in d2:
            # 顯示修改前的舊資料
            print(f'Current content -> {key}: {d2[key]}')
            value = input('Enter a new value (String): ')
            # 覆蓋原有的 Value
            d2[key] = value
            print("Modification successful.")
        else:
            print(f'Key {key} not found.')
    except ValueError:
        print("Invalid input.")
```

```
def display(d2):
    """列出目前字典的所有資料狀態"""
    print(f"Current Dictionary: {d2}")
```

```
def main():
    """主控迴圈，驅動整個管理系統邏輯"""
    d = {} # 初始化資料容器
    while True:
        choice = menu()
        if choice == 1:
            add(d)
        elif choice == 2:
            delete(d)
        elif choice == 3:
            modify(d)
        elif choice == 4:
            display(d)
        elif choice == 5:
            print("Goodbye!")
            break
        else:
            print('Invalid choice, please try again.')
```

```
if __name__ == "__main__":
    main()
```
