```
"""
檔名：Text_File_Unique_Word_Sorter.py
功能：進階資料結構練習
學習重點：不可變與可變資料類型、集合運算
"""
def main():
    # 取得使用者輸入的檔名並移除前後空白
    f1 = input("Enter a filename: ").strip()

    # 開啟檔案進行讀取模式
    infile = open(f1, "r")
    # 讀取檔案全部內容
    s = infile.read()
    print(f"{f1} contains following words: {s}")
    # 讀取完畢後關閉檔案流
    infile.close()

    # 將字串內容以空白字元切割成單字列表 (List)
    words = s.split()
    
    # 利用 set 集合「元素唯一性」的特性，快速過濾掉重複的單字
    nonduplicate_words = set(words)
    
    # 將集合轉回列表，因為集合本身是無序的，轉回列表才能進行排序
    words = list(nonduplicate_words)
    
    # 針對列表進行原地排序 (預設為 A-Z 字母順序)
    words.sort()

    print('After sorted...')
    # 遍歷排序後的列表並印出，使用 end='' 保持在同一行輸出
    for word in words:
        print(word, end=' ')
```

```
# 執行主程式
if __name__ == "__main__":
    main()
```
