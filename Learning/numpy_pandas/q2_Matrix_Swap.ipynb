```
"""
檔名：q2_Matrix_Swap.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import numpy as np
```

```
# 產生 1~200 偶數的 10x10 矩陣
a = np.arange(2, 201, 2).reshape(10, 10)
```

```
# 產生 1~200 奇數的 10x10 矩陣
b = np.arange(1, 200, 2).reshape(10, 10)
```

```
print("交換前：")
print(f"a:\n{a}\n")
print(f"b:\n{b}\n")
```

```
# 將兩矩陣相互交換（全部交換）
a, b = b.copy(), a.copy()
```

```
print("交換後：")
print(f"a:\n{a}\n")
print(f"b:\n{b}")
```
