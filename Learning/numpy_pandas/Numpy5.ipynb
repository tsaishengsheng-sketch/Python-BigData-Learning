```
"""
檔名：Numpy5.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import numpy as np
```

```
a = np.array([30, 65, 21, 90, 27, 43, 67, 60, 87, 72])
b = np.array([52, 91, 38, 22, 27, 60])
c = np.arange(5, 96, 10)  # [5, 15, 25, 35, 45, 55, 65, 75, 85, 95]
```

```
# a 和 b 的交集（兩個都有的）
intersection = np.intersect1d(a, b)
```

```
# a 和 c 的差集（a 有但 c 沒有的）
difference = np.setdiff1d(a, c)
```

```
# a 和 b 的聯集（合併去重複）
union = np.union1d(a, b)
```

```
print(f"交集結果：{intersection}")
print(f"差集結果：{difference}")
print(f"聯集結果：{union}")
```
