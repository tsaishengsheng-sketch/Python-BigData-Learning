```
"""
檔名：Numpy4.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import numpy as np
```

```
# 產生 1~50 之間奇數的 5x5 矩陣（1, 3, 5 ... 49）
a = np.arange(1, 50, 2).reshape(5, 5)
```

```
# 產生 1~50 之間偶數的 5x5 矩陣（2, 4, 6 ... 50）
b = np.arange(2, 51, 2).reshape(5, 5)
```

```
# 垂直堆疊：a 疊在上面，b 疊在下面，結果變 10x5
v_stack = np.concatenate([a, b], axis=0)
```

```
# 水平堆疊：a 在左邊，b 在右邊，結果變 5x10
h_stack = np.concatenate([a, b], axis=1)
```

```
print(f"垂直堆疊：\n{v_stack}")
print()
print(f"水平堆疊：\n{h_stack}")
```
