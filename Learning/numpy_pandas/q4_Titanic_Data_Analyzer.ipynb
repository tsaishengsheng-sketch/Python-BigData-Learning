```
"""
檔名：q4_Titanic_Data_Analyzer.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import pandas as pd
import pydataset
```

```
# 載入 Titanic 資料集
df = pd.DataFrame(pydataset.data("Titanic"))
```

```
# 去掉 Class 和 Freq 欄位
df = df.drop(columns=["Class", "Freq"])
```

```
# 將欄位名稱改為中文
df = df.rename(columns={"Sex": "性別", "Age": "年齡", "Survived": "是否存活"})
```

```
print(f"原始資料（前10筆）：\n{df.head(10)}\n")
```

```
# 存活的男性
print(f"存活的男性：\n{df[(df['性別'] == 'Male') & (df['是否存活'] == 'Yes')]}")
```
