```
"""
檔名：pandas2.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import pandas as pd
```

```
# 商品名稱
products = ["iPhone", "Samsung", "Sony", "Huawei", "HTC", "Oppo"]
# 存貨量
inventory = [1000, 1200, 1300, 800, 700, 850]
```

```
# 建立字典，以 data 的 keys 作為欄位名稱
data = {"商品名稱": products, "存貨": inventory}
df = pd.DataFrame(data, columns=data.keys())
```

```
# 將 iPhone 的存貨加50
df.loc[0, "存貨"] += 50
```

```
# 輸出存貨最多的商品
print(f"存貨最多的商品：\n{df[df['存貨'] == df['存貨'].max()]}")
```

```
# 將資料根據存貨量從大到小排序
print(f"\n根據存貨從大到小排序：\n{df.sort_values(by='存貨', ascending=False)}")
```

```
# 輸出存貨量少於1000的商品
print(f"\n存貨量少於1000的商品：\n{df[df['存貨'] < 1000]}")
```
