```
"""
檔名：pandas3.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import pandas as pd
```

```
# 建立名單--學生名字
names = ["Cathy", "Daisy", "Bonnie", "Vicky", "Mary"]
# 建立名單--學生分數
scores = [84, 98, 92, 80, 75]
```

```
# 建立字典並轉成 DataFrame
data = {"Name": names, "Score": scores}
df = pd.DataFrame(data)
```

```
# 原始資料
print(f"原始資料：\n{df}\n")
```

```
# 根據名字從小到大排序
df = df.sort_values(by="Name")
print(f"將資料根據名字從小到大排序：\n{df}\n")
```

```
# 根據分數從大到小排序
df = df.sort_values(by="Score", ascending=False)
print(f"將資料根據分數從大到小排序：\n{df}\n")
```

```
# 列出分數高於90分的學生
greater90 = df[df["Score"] > 90]
print(f"列出分數高於90分的學生：\n{greater90}")
```
