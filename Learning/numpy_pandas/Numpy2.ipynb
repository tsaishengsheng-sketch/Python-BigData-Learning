```
"""
檔名：Numpy2.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import numpy as np
```

```
# 固定亂數種子，確保每次結果相同
np.random.seed(42)
```

```
# 產生 3x4 的矩陣（12個數字，範圍 1~50）
matrix1 = np.random.randint(1, 51, 12).reshape(3, 4)
```

```
# 產生 4x5 的矩陣（20個數字，範圍 1~50）
matrix2 = np.random.randint(1, 51, 20).reshape(4, 5)
```

```
# 顯示兩個矩陣
print(f"matrix1:\n{matrix1}")
print(f"\nmatrix2:\n{matrix2}")
```

```
# 每一列的最大值（axis=1 = 橫的比，每列找最大）
print(f"\nmatrix1 每一列的最大值：{np.amax(matrix1, axis=1)}")
```

```
# matrix1 第2列小於30的個數
print(f"\nmatrix1 第2列小於30的個數：{np.sum(matrix1[2, :] < 30)}")
```

```
# matrix2 每一欄的最大值（axis=0 = 直的比，每欄找最大）
print(f"\nmatrix2 每一欄的最大值：{np.amax(matrix2, axis=0)}")
```

```
# matrix2 第2欄小於30的個數
print(f"\nmatrix2 第2欄小於30的個數：{np.sum(matrix2[:, 2] < 30)}")
```

```
# matrix1 第一列和 matrix2 第一列的聯集（合併去重複）
print(
    f"\nmatrix1 第一列和 matrix2 第一列的聯集結果：{np.union1d(matrix1[0, :], matrix2[0, :])}"
)
```

```
# matrix1 * matrix2 矩陣相乘（3x4 乘以 4x5 = 3x5）
print(f"\nmatrix1 * matrix2:\n{np.dot(matrix1, matrix2)}")
```
