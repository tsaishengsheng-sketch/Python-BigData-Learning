```
"""
檔名：Numpy1.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import numpy as np
```

```
# 固定亂數種子：讓每次產生的資料都一樣，方便對照結果
# 數字 42 只是慣例，可以換成任何整數
np.random.seed(42)
```

```
# 隨機產生 80 個 1~100 的整數
data = np.random.randint(1, 101, size=80)
```

```
# 印出原始資料，讓人知道在算什麼
print(f"原始資料：\n{data}\n")
```

```
# 印出資料型態（會顯示 numpy.ndarray）
print(f"資料型態：{type(data)}")
```

```
# 平均值：所有數字加總除以個數
print(f"平均值：{np.mean(data):.2f}")
```

```
# 中位數：排序後位於中間的值
print(f"中位數：{np.median(data):.2f}")
```

```
# 標準差：資料離平均值的平均距離，越大代表越分散
print(f"標準差：{np.std(data):.2f}")
```

```
# 變異數：標準差的平方
print(f"變異數：{np.var(data):.2f}")
```

```
# 極差值：最大值 - 最小值
print(f"極差值：{np.ptp(data):.2f}")
```
