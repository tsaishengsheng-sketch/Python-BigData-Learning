```
"""
檔名：q3_Student_Score_Analyzer.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import pandas as pd
```

```
names = ["Nancy", "Daisy", "Bonnie", "Vicky", "John"]
scores = [84, 98, 92, 82, 78]
```

```
data = {"Name": names, "Score": scores}
df = pd.DataFrame(data)
```

```
# 原始資料
print(f"原始資料：\n{df}\n")
```

```
# 根據名字從小到大排序
df = df.sort_values(by="Name")
print(f"根據名字從小到大排序：\n{df}\n")
```

```
# 根據分數從大到小排序
df = df.sort_values(by="Score", ascending=False)
print(f"根據分數從大到小排序：\n{df}\n")
```

```
# 分數高於90分的學生
print(f"分數高於90分的學生：\n{df[df['Score'] > 90]}")
```
