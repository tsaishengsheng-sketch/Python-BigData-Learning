```
"""
檔名：pandas1.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import pandas as pd
import numpy as np
import os
```

```
# 讀取 csv 檔案並轉成 DataFrame
df = pd.read_csv(os.path.join(os.path.dirname(__file__), "file1.csv"))
```

```
# 改變索引為從 1 開始並顯示結果
df = df.set_index(np.arange(1, len(df) + 1))
print(df)
```

```
# 將進貨日期由字串形式轉換為數值形式（方便之後排序）
df["進貨日期"] = pd.to_datetime(df["進貨日期"], format="%Y/%m/%d")
```

```
# 顯示前兩筆記錄
print(f"\n前兩筆記錄：\n{df[:2]}")
```

```
# 輸出白醋的進貨記錄
print(f"\n'白醋'的進貨記錄：\n{df[df['商品名稱'] == '白醋']}")
```

```
# 計算各單位商品數量並輸出
print(f"\n各單位商品數量：\n{df.groupby('單位')['商品名稱'].count()}")
```

```
# 根據進貨日期從小到大排序並輸出結果
print(f"\n根據進貨日期從小到大排序：\n{df.sort_values(by='進貨日期')}")
```
