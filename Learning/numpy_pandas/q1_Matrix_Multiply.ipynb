```
"""
檔名：q1_Matrix_Multiply.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import numpy as np
```

```
# 產生 1 到 9 的 3x3 矩陣
matrix1 = np.arange(1, 10).reshape(3, 3)
```

```
# 產生 9 到 1 的 3x3 矩陣
matrix2 = np.arange(9, 0, -1).reshape(3, 3)
```

```
print(f"matrix1:\n{matrix1}\n")
print(f"matrix2:\n{matrix2}\n")
```

```
# 兩矩陣相乘
print(f"matrix1 * matrix2:\n{np.dot(matrix1, matrix2)}")
```
