```
"""
檔名：Numpy3.py
功能：資料分析練習
學習重點：NumPy 陣列運算、Pandas DataFrame 操作
"""
import numpy as np
```

```
# 建立 1~200 之間偶數的 10x10 矩陣（2, 4, 6 ... 200）
a = np.arange(2, 201, 2).reshape(10, 10)
```

```
# 建立 1~200 之間奇數的 10x10 矩陣（1, 3, 5 ... 199）
b = np.arange(1, 200, 2).reshape(10, 10)
```

```
print("交換前：")
print("a:")
print(a)
print("b:")
print(b)
```

```
print()
```

```
# 將兩個矩陣的偶數列作交換（index 1, 3, 5, 7, 9 = 第2, 4, 6, 8, 10列）
a[range(1, 10, 2)], b[range(1, 10, 2)] = (
    b[range(1, 10, 2)].copy(),
    a[range(1, 10, 2)].copy(),
)
```

```
print("交換後：")
print("a:")
print(a)
print("b:")
print(b)
```
