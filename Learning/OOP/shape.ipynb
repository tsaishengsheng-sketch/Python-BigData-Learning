```
"""
檔名：shape.py
功能：物件導向程式設計
學習重點：類別定義、物件實例化與封裝應用
"""
import math
class Shape:
    def __init__(self, xPoint =0, yPoint =0):
        self.xPoint = xPoint
        self.yPoint = yPoint
    
    def getPoint(self):
        return (self.xPoint, self.yPoint)
    
    def setPoint(self, xPoint, yPoint):
        self.xPoint = xPoint
        self.yPoint = yPoint
    
    def __str__(self):
        print('xPoint = %d, yPoint = %d' % (self.xPoint, self.yPoint))
```
