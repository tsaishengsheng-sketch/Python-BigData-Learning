```
import math  # 用來計算平方根（海龍公式求面積）
```

```
# ────────────────────────────────────────
# 父類別：Shape（形狀）
# ────────────────────────────────────────
class Shape:
    """所有形狀的基底類別，擁有顏色屬性"""

    def __init__(self, color="白色"):
        # __color 是 private 屬性，子類別也無法直接存取，需透過 getter/setter
        self.__color = color

    # Getter：取得顏色
    def getColor(self):
        return self.__color

    # Setter：設定顏色
    def setColor(self, color):
        self.__color = color

    # 印出顏色
    def printColor(self):
        print(f"  顏色：{self.__color}")
```

```
# ────────────────────────────────────────
# 子類別：Triangle（三角形），繼承自 Shape
# ────────────────────────────────────────
class Triangle(Shape):
    """三角形類別，繼承 Shape，新增三邊長相關功能"""

    def __init__(self, color="白色", a=0, b=0, c=0):
        # 呼叫父類別的 __init__，設定顏色
        super().__init__(color)
        # 三條邊都是 private 屬性
        self.__a = a
        self.__b = b
        self.__c = c

    # ── Getter：取得三邊 ──
    def getSideA(self):
        return self.__a

    def getSideB(self):
        return self.__b

    def getSideC(self):
        return self.__c

    # ── Setter：設定三邊 ──
    def setSideA(self, a):
        self.__a = a

    def setSideB(self, b):
        self.__b = b

    def setSideC(self, c):
        self.__c = c

    def setSides(self, a, b, c):
        """一次設定三邊"""
        self.__a = a
        self.__b = b
        self.__c = c

    def perimeter(self):
        """計算周長：三邊相加"""
        return self.__a + self.__b + self.__c

    def area(self):
        """
        計算面積：使用海龍公式 (Heron's Formula)
        s = 半周長，area = √(s(s-a)(s-b)(s-c))
        """
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.__a) * (s - self.__b) * (s - self.__c))

    def printInfo(self):
        """印出三角形的完整資訊（顏色 + 三邊長）"""
        self.printColor()  # 呼叫繼承自 Shape 的函式
        print(f"  三邊長：a={self.__a}, b={self.__b}, c={self.__c}")
        print(f"  周長：{self.perimeter()}")
        print(f"  面積：{self.area():.2f}")
```

```
# ──────────────── 主程式測試 ────────────────
if __name__ == "__main__":
    print("=== Triangle 三角形測試 ===\n")

    # 建立物件，傳入顏色與三邊長
    t = Triangle("紅色", 3, 4, 5)

    print("【直接建立物件，印出資訊】")
    t.printInfo()

    print("\n【用 setter 修改顏色與邊長】")
    t.setColor("藍色")
    t.setSides(6, 8, 10)
    t.printInfo()
```
