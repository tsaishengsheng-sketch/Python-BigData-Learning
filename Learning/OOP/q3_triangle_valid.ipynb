```
"""
檔名：q3_triangle_valid.py
功能：物件導向程式設計
學習重點：類別定義、物件實例化與封裝應用
"""
# ============================================================
# 第3題：承第2題，加入三角形合法性驗證
# 三角形成立條件：任意兩邊之和 > 第三邊
# 不合法時印出提示訊息，不做計算
# ============================================================
```

```
import math
```

```
class Shape:
    """父類別：形狀"""

    def __init__(self, color="白色"):
        self.__color = color

    def getColor(self):
        return self.__color

    def setColor(self, color):
        self.__color = color

    def printColor(self):
        print(f"  顏色：{self.__color}")
```

```
class Triangle(Shape):
    """三角形類別（含合法性驗證）"""

    def __init__(self, color="白色", a=0, b=0, c=0):
        super().__init__(color)
        self.__a = a
        self.__b = b
        self.__c = c

    def getSideA(self):
        return self.__a

    def getSideB(self):
        return self.__b

    def getSideC(self):
        return self.__c

    def setSides(self, a, b, c):
        self.__a = a
        self.__b = b
        self.__c = c

    def isValid(self):
        """
        驗證三邊能否構成三角形
        條件：每一邊都必須 > 0，且任意兩邊之和 > 第三邊
        回傳 True（合法）或 False（不合法）
        """
        a, b, c = self.__a, self.__b, self.__c

        # 先確認每一邊都是正數
        if a <= 0 or b <= 0 or c <= 0:
            return False

        # 三角不等式：任意兩邊之和必須大於第三邊
        return (a + b > c) and (a + c > b) and (b + c > a)

    def perimeter(self):
        """計算周長（呼叫前請先確認 isValid()）"""
        return self.__a + self.__b + self.__c

    def area(self):
        """計算面積：海龍公式"""
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.__a) * (s - self.__b) * (s - self.__c))

    def printInfo(self):
        """印出資訊前先驗證，不合法則印出警告"""
        a, b, c = self.__a, self.__b, self.__c
        print(f"  輸入三邊：a={a}, b={b}, c={c}")

        if not self.isValid():
            # 不合法：說明原因
            print("  ✗ 此三邊無法構成三角形！（任意兩邊之和必須大於第三邊）")
            return  # 提早結束，不繼續計算

        # 合法才印出詳細資料
        self.printColor()
        print(f"  周長：{self.perimeter()}")
        print(f"  面積：{self.area():.2f}")
```

```
# ──────────────── 主程式測試 ────────────────
if __name__ == "__main__":
    print("=== Triangle 三角形驗證測試 ===\n")

    print("【測試1：合法三角形 (3, 4, 5)】")
    t1 = Triangle("綠色", 3, 4, 5)
    t1.printInfo()

    print("\n【測試2：不合法三角形 (1, 2, 10)】")
    t2 = Triangle("黃色", 1, 2, 10)
    t2.printInfo()

    print("\n【測試3：退化三角形，兩邊之和等於第三邊 (1, 2, 3)】")
    t3 = Triangle("橘色", 1, 2, 3)
    t3.printInfo()

    print("\n【測試4：等邊三角形 (5, 5, 5)】")
    t4 = Triangle("紫色", 5, 5, 5)
    t4.printInfo()
```
