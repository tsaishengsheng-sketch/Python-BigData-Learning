```
def main():
    # 創建一個 Lion 物件
    lionObj = Lion('Luke')
    print('Lion Object')
    print(f'Name: {lionObj.getName()}')
    print(f'Breed: {lionObj.breed()}')
    print(f'Food: {lionObj.food()}')
    print()
    
    # 創建一個 Duck 物件
    duckObj = Duck('Kiki')
    print('Duck Object')
    print(f'Name: {duckObj.getName()}')
    print(f'Breed: {duckObj.breed()}')
    print(f'Food: {duckObj.food()}')
    print()

    # 創建一個 Duck 物件
    sheepobj = Sheep('Nala')
    print ('sheep object')
    print(f'Name: {sheepobj.getName()}')
    print(f'Breed: {sheepobj.breed()}')
    print(f'Food: {sheepobj.food()}')
    print()
```
