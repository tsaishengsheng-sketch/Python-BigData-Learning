```
class Stack:
    """堆疊類別：後進先出 (Last In, First Out)"""

    def __init__(self):
        # __items 是 private 屬性（名稱前加雙底線），外部無法直接存取
        self.__items = []

    def push(self, item):
        """將一個元素加入堆疊頂端"""
        self.__items.append(item)
        print(f"  → 推入 '{item}'，目前堆疊：{self.__items}")

    def pop(self):
        """移除並回傳堆疊頂端的元素"""
        if self.isEmpty():
            print("  ✗ 堆疊是空的，無法 pop！")
            return None
        item = self.__items.pop()  # list.pop() 預設取出最後一個元素
        print(f"  ← 彈出 '{item}'，目前堆疊：{self.__items}")
        return item

    def isEmpty(self):
        """判斷堆疊是否為空，回傳 True / False"""
        return len(self.__items) == 0

    def getSize(self):
        """回傳堆疊目前的大小（元素數量）"""
        return len(self.__items)
```

```
# ──────────────── 主程式測試 ────────────────
if __name__ == "__main__":
    print("=== Stack 堆疊測試 ===\n")

    s = Stack()

    print("【測試 isEmpty】")
    print(f"  堆疊是否為空？{s.isEmpty()}")  # True

    print("\n【測試 push：依序推入 10, 20, 30】")
    s.push(10)
    s.push(20)
    s.push(30)

    print(f"\n  堆疊大小：{s.getSize()}")  # 3

    print("\n【測試 pop：彈出兩次】")
    s.pop()  # 取出 30（最後推入的）
    s.pop()  # 取出 20

    print(f"\n  堆疊大小：{s.getSize()}")  # 1

    print("\n【連續 pop 到底，測試空堆疊保護】")
    s.pop()  # 取出 10
    s.pop()  # 應提示「堆疊是空的」
```
