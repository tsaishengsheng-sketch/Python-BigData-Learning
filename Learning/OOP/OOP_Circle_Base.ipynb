```python
"""
檔名：OOP_Circle_Base.py
功能：物件導向程式設計
學習重點：類別定義、物件實例化與封裝應用
"""
```

```python
import math
class Circle:
    def __init__(self, radius =1):
        self.radius = radius

    def setRadius(self, radius):
        self.radius = radius
    
    def getPerimeter(self):
        perimeter = 2 * math.pi * self.radius
        return perimeter
    
    def getArea(self):
        area = math.pi * self.radius ** 2
        return area
```

```python
def main():
    circle1 = Circle()
    print(f"circle1: radius is {circle1.radius}, perimeter is {circle1.getPerimeter():.2f}, area is {circle1.getArea():.2f}")

    print()
    circle2 = Circle(5)
    print(f"circle2: radius is {circle2.radius}, perimeter is {circle2.getPerimeter():.2f}, area is {circle2.getArea():.2f}")
```

```python
main()
```
