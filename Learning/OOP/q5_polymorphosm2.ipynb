```
"""
檔名：q5_polymorphosm2.py
功能：物件導向程式設計
學習重點：類別定義、物件實例化與封裝應用
"""
# ============================================================
# 第5題：polymorphism2.py
# 在原本的多型範例中，加入 Triangle（三角形）類別
#
# 多型 (Polymorphism) 概念：
#   - 不同的子類別都有同一個方法名稱（例如 area()、printInfo()）
#   - 呼叫時不需要知道物件是哪種形狀，統一用相同方法呼叫
#   - Python 透過「鴨子定型 (Duck Typing)」自然支援多型
# ============================================================
```

```
import math
```

```
# ────────────────────────────────────────
# 自訂異常（從第4題延續）
# ────────────────────────────────────────
class InvalidTriangleError(Exception):
    """三邊無法構成三角形時拋出"""

    pass
```

```
# ────────────────────────────────────────
# 父類別：Shape
# ────────────────────────────────────────
class Shape:
    """所有形狀的基底類別"""

    def __init__(self, color="白色"):
        self.__color = color

    def getColor(self):
        return self.__color

    def setColor(self, color):
        self.__color = color

    def area(self):
        """面積：子類別必須覆寫 (override) 這個方法"""
        return 0

    def perimeter(self):
        """周長：子類別必須覆寫這個方法"""
        return 0

    def printInfo(self):
        """印出形狀資訊"""
        print(f"  形狀：{self.__class__.__name__}")
        print(f"  顏色：{self.__color}")
        print(f"  面積：{self.area():.2f}")
        print(f"  周長：{self.perimeter():.2f}")
```

```
# ────────────────────────────────────────
# 子類別：Circle（圓形）
# ────────────────────────────────────────
class Circle(Shape):
    """圓形類別"""

    def __init__(self, color="白色", radius=0):
        super().__init__(color)
        self.__radius = radius  # 半徑

    def getRadius(self):
        return self.__radius

    def setRadius(self, r):
        self.__radius = r

    def area(self):
        """圓面積 = π × r²"""
        return math.pi * self.__radius**2

    def perimeter(self):
        """圓周長 = 2 × π × r"""
        return 2 * math.pi * self.__radius
```

```
# ────────────────────────────────────────
# 子類別：Rectangle（長方形）
# ────────────────────────────────────────
class Rectangle(Shape):
    """長方形類別"""

    def __init__(self, color="白色", width=0, height=0):
        super().__init__(color)
        self.__width = width
        self.__height = height

    def area(self):
        """長方形面積 = 寬 × 高"""
        return self.__width * self.__height

    def perimeter(self):
        """長方形周長 = 2 × (寬 + 高)"""
        return 2 * (self.__width + self.__height)
```

```
# ────────────────────────────────────────
# 子類別：Triangle（三角形）← 第5題新增
# ────────────────────────────────────────
class Triangle(Shape):
    """三角形類別（含異常驗證）"""

    def __init__(self, color="白色", a=0, b=0, c=0):
        super().__init__(color)
        self.__validateAndSet(a, b, c)

    def __validateAndSet(self, a, b, c):
        if a <= 0 or b <= 0 or c <= 0:
            raise InvalidTriangleError(f"三邊必須為正數：a={a}, b={b}, c={c}")
        if not (a + b > c and a + c > b and b + c > a):
            raise InvalidTriangleError(f"三邊 ({a},{b},{c}) 無法構成三角形！")
        self.__a = a
        self.__b = b
        self.__c = c

    def area(self):
        """海龍公式計算面積"""
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.__a) * (s - self.__b) * (s - self.__c))

    def perimeter(self):
        """三邊相加"""
        return self.__a + self.__b + self.__c
```

```
# ──────────────── 主程式測試（展示多型） ────────────────
if __name__ == "__main__":
    print("=== 多型 (Polymorphism) 展示 ===\n")

    # 建立不同種類的形狀物件，放進同一個串列
    shapes = [
        Circle("紅色", 5),
        Rectangle("藍色", 4, 6),
        Triangle("綠色", 3, 4, 5),  # ← 第5題新增的三角形
        Triangle("橘色", 5, 12, 13),  # ← 另一個直角三角形
    ]

    # 用同一個迴圈處理所有形狀（多型的精髓！）
    # 不需要判斷每個物件是什麼類型，直接呼叫 printInfo()
    for i, shape in enumerate(shapes, 1):
        print(f"── 形狀 {i} ──")
        shape.printInfo()  # 每個子類別自動呼叫自己的 area() 和 perimeter()
        print()

    # ── 也可以統計各形狀面積總和 ──
    print("── 各形狀面積總計 ──")
    total = sum(shape.area() for shape in shapes)
    print(f"  全部面積加總：{total:.2f}")

    # ── 測試異常處理（不合法三角形加入時的保護）──
    print("\n── 嘗試加入不合法三角形 ──")
    try:
        bad_triangle = Triangle("黑色", 1, 2, 100)
    except InvalidTriangleError as e:
        print(f"  ✗ 異常攔截：{e}")
        print("  （其餘形狀不受影響）")
```
