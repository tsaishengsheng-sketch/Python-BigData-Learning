```
"""
檔名：isinstance_example.py
功能：物件導向程式設計
學習重點：類別定義、物件實例化與封裝應用
"""
from shape import Circle, Rectangle
```

```
def displayInform(obj):
    """ 輸出形狀的信息 """
    print(f"Area: {obj.getArea():.2f}")
    if isinstance(obj, Circle):
        print(f"Perimeter: {obj.getPerimeter():.2f}")
        print(f"Radius: {obj.getRadius()}")
    elif isinstance(obj, Rectangle):
        print(f"Perimeter: {obj.getPerimeter():.2f}")
        print(f"Width: {obj.getWidth()}")
        print(f"Height: {obj.getHeight()}")
```

```
def main():
    # 創建一個 Circle 物件並輸出其信息
    circle = Circle(5)
    print('Circle information')
    displayInform(circle)
    print()
    
    # 創建一個 Rectangle 物件並輸出其信息
    rectangle = Rectangle(3, 4)
    print('Rectangle information')
    displayInform(rectangle)
    print()
```

```
# 調用主函數
if __name__ == "__main__":
    main()
```
