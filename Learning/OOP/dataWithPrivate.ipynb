```
"""
檔名：dataWithPrivate.py
功能：物件導向程式設計
學習重點：類別定義、物件實例化與封裝應用
"""
import math
class Circle:
    def __init__(self, radius =1):
        self.__radius = radius

    def setRadius(self, radius):
        self.__radius = radius
    
    def getRadius(self):
        return self.__radius
    
    def getPerimeter(self):
        perimeter = 2 * math.pi * self.__radius
        return perimeter
    
    def getArea(self):
        area = math.pi * self.__radius ** 2
        return area
```

```
def main():
    circle1 = Circle()
    print(f"circle1: radius is {circle1.getRadius()}, perimeter is {circle1.getPerimeter():.2f}, area is {circle1.getArea():.2f}")

    print()
    circle2 = Circle(5)
    print(f"circle2: radius is {circle2.getRadius()}, perimeter is {circle2.getPerimeter():.2f}, area is {circle2.getArea():.2f}")
```

```
main()
```
