```
"""
檔名：BMI.py
功能：物件導向程式設計
學習重點：類別定義、物件實例化與封裝應用
"""
class Bmi:
    def __init__(self, weight=170, height=60):
        """ 初始化 Bmi 物件，設置初始體重和身高 """
        self._weight = weight
        self.height = height
    
    def getBmi(self):
        """ 計算並輸出 BMI 指數及其狀態 """
        heightMeter = self.height / 100
        bmi = self._weight / (heightMeter * heightMeter)
        print(f"BMI: {bmi:.2f}")
        
        if bmi < 18.5:
            print("Underweight")
        elif bmi < 25:
            print("Normal")
        elif bmi < 30:
            print("Overweight")
        else:
            print("Obesity")
```

```
def main():
    # 創建一個 Bmi 物件 John
    John = Bmi(68, 185)
    print("John's BMI is:", end=" ")
    John.getBmi()
    print()
    
    # 創建一個 Bmi 物件 Mary
    Mary = Bmi(53, 172)
    print("Mary's BMI is:", end=" ")
    Mary.getBmi()
    print()
```

```
# 調用主函數
if __name__ == "__main__":
    main()
```
