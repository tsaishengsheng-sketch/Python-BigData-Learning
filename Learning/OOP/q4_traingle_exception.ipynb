```
"""
檔名：q4_traingle_exception.py
功能：物件導向程式設計
學習重點：類別定義、物件實例化與封裝應用
"""
# ============================================================
# 第4題：承第3題，改用「異常處理 (Exception)」處理不合法三角形
# 異常處理概念：
#   - raise：主動拋出一個錯誤
#   - try / except：嘗試執行，若發生錯誤就攔截處理
#   - 自訂 Exception：建立自己專屬的錯誤類別，訊息更清楚
# ============================================================
```

```
import math
```

```
# ────────────────────────────────────────
# 自訂異常類別（繼承內建的 Exception）
# 這樣我們就有一個專屬於「三角形錯誤」的例外類型
# ────────────────────────────────────────
class InvalidTriangleError(Exception):
    """當三邊無法構成三角形時，拋出此異常"""

    pass  # 不需要額外內容，繼承 Exception 即可
```

```
class Shape:
    """父類別：形狀"""

    def __init__(self, color="白色"):
        self.__color = color

    def getColor(self):
        return self.__color

    def setColor(self, color):
        self.__color = color

    def printColor(self):
        print(f"  顏色：{self.__color}")
```

```
class Triangle(Shape):
    """三角形類別（使用異常處理驗證三邊）"""

    def __init__(self, color="白色", a=0, b=0, c=0):
        super().__init__(color)
        # 在建立物件時就立刻驗證，不合法直接拋出異常
        self.__validateAndSet(a, b, c)

    def __validateAndSet(self, a, b, c):
        """
        private 函式：驗證三邊並存入屬性
        不合法就 raise InvalidTriangleError
        """
        if a <= 0 or b <= 0 or c <= 0:
            raise InvalidTriangleError(f"三邊長必須為正數！輸入值：a={a}, b={b}, c={c}")
        if not (a + b > c and a + c > b and b + c > a):
            raise InvalidTriangleError(
                f"三邊 ({a}, {b}, {c}) 無法構成三角形！（任意兩邊之和必須大於第三邊）"
            )
        # 驗證通過才存入
        self.__a = a
        self.__b = b
        self.__c = c

    def setSides(self, a, b, c):
        """重新設定三邊（同樣會做驗證）"""
        self.__validateAndSet(a, b, c)

    def perimeter(self):
        return self.__a + self.__b + self.__c

    def area(self):
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.__a) * (s - self.__b) * (s - self.__c))

    def printInfo(self):
        self.printColor()
        print(f"  三邊長：a={self.__a}, b={self.__b}, c={self.__c}")
        print(f"  周長：{self.perimeter()}")
        print(f"  面積：{self.area():.2f}")
```

```
# ──────────────── 主程式測試 ────────────────
if __name__ == "__main__":
    print("=== Triangle 異常處理測試 ===\n")

    # ── 測試1：合法三角形，正常建立 ──
    print("【測試1：合法三角形 (3, 4, 5)】")
    try:
        t1 = Triangle("藍色", 3, 4, 5)
        t1.printInfo()
    except InvalidTriangleError as e:
        print(f"  ✗ 異常攔截：{e}")

    # ── 測試2：不合法三角形，觸發異常 ──
    print("\n【測試2：不合法三角形 (1, 2, 10)】")
    try:
        t2 = Triangle("紅色", 1, 2, 10)  # 這行會拋出異常
        t2.printInfo()  # 這行不會被執行到
    except InvalidTriangleError as e:
        # except 攔截到異常，印出錯誤訊息，程式不會崩潰
        print(f"  ✗ 異常攔截：{e}")

    # ── 測試3：邊長為負數 ──
    print("\n【測試3：邊長為負數 (-1, 2, 3)】")
    try:
        t3 = Triangle("綠色", -1, 2, 3)
        t3.printInfo()
    except InvalidTriangleError as e:
        print(f"  ✗ 異常攔截：{e}")

    # ── 測試4：用 setSides 重新設定，也會驗證 ──
    print("\n【測試4：先建立合法三角形，再用 setSides 設定不合法值】")
    try:
        t4 = Triangle("橘色", 5, 5, 5)
        print("  建立成功：(5, 5, 5)")
        t4.setSides(0, 1, 1)  # 這行會拋出異常
    except InvalidTriangleError as e:
        print(f"  ✗ 異常攔截：{e}")

    print("\n程式正常結束（異常都被妥善處理了！）")
```
