```
"""
檔名：animalClass.py
功能：物件導向程式設計
學習重點：類別定義、物件實例化與封裝應用
"""
class Animal:
    def __init__(self, name='Unknown'):
        """ 初始化 Animal 物件，設置初始名稱 """
        self._name = name
    
    def setName(self, name):
        """ 設置新的名稱 """
        self._name = name
    
    def getName(self):
        """ 返回當前的名稱 """
        return self._name
```

```
class Lion(Animal):
    def __init__(self, name):
        """ 初始化 Lion 物件，設置初始名稱 """
        super().__init__(name)
    
    def breed(self):
        """ 輸出胎生狀態 """
        return 'viviparous'
    
    def food(self):
        """ 輸出 """
        return 'meat'
```

```
class Duck(Animal):
    def __init__(self, name):
        """ 初始化 Duck 物件，設置初始名稱 """
        super().__init__(name)
    
    def breed(self):
        """ 輸出卵生狀態 """
        return 'Oviparous'
    
    def food(self):
        """ 輸出 """
        return 'grass'
```
