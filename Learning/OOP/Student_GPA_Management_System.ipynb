```
"""
檔名：Student_GPA_Management_System.py
功能：物件導向程式設計
學習重點：類別定義、物件實例化與封裝應用
"""
class Gpa:
    # 1. 初始化方法 (__init__)：物件「開箱」時設定初始資料
    def __init__(self, name = 'Nancy', score = 60):
        self.name = name    # 把傳入的名字存進物件口袋
        self.score = score  # 把傳入的分數存進物件口袋

    # 2. 修改方法 (Setter)：物件出生後，用來強行換掉口袋裡的名字
    def setName(self, name):
        self.name = name

    # 3. 修改方法 (Setter)：物件出生後，用來強行換掉口袋裡的分數
    def setScore(self, score):
        self.score = score

    # 4. 邏輯判斷方法：根據物件「自己口袋裡的分數」印出結果
    def getGpa(self):
        # 注意：此處符號建議修正為 >= 以符合一般邏輯
        if self.score >= 80:
            print(f"{self.name} 的 GPA 為 A")
        elif self.score >= 70:
            print(f"{self.name} 的 GPA 為 B")  
        elif self.score >= 60:
            print(f"{self.name} 的 GPA 為 C")
        elif self.score >= 50:
            print(f"{self.name} 的 GPA 為 D")
        else:
            print(f"{self.name} 的 GPA 為 E")
```

```
def main():
    # 情況 A：不傳參數，使用預設值 (Nancy)
    gpa0 = Gpa()
    gpa0.getGpa()
    
    # 情況 B：創建物件時直接傳入資料，這就是為什麼能印出 John
    gpa1 = Gpa('John', 82)
    gpa1.getGpa()
    
    # 情況 C：同上，傳入 Mary
    gpa2 = Gpa('Mary', 75)
    gpa2.getGpa()
    
    # 情況 D：先用預設值創建物件，再透過 Setter 方法修改成 Tom
    gpa3 = Gpa()
    gpa3.setName('Tom')
    gpa3.setScore(65)
    gpa3.getGpa()   
```

```
# 呼叫主程式
if __name__ == "__main__":
    main()
```
