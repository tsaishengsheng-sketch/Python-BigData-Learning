```
from shape import Circle, Rectangle
```

```
def displayArea(obj):
    """ 輸出形狀的面積 """
    print(f"Area: {obj.getArea():.2f}")
```

```
def displayPerimeter(obj):
    """ 輸出形狀的周長 """
    print(f"Perimeter: {obj.getPerimeter():.2f}")
```

```
def main():
    # 創建一個 Circle 物件並輸出其信息
    circle = Circle(5)
    print(circle)
    displayPerimeter(circle)
    displayArea(circle)
    print()
    
    # 創建一個 Rectangle 物件並輸出其信息
    rectangle = Rectangle(3, 4)
    print(rectangle)
    displayPerimeter(rectangle)
    displayArea(rectangle)
    print()
```

```
# 調用主函數
if __name__ == "__main__":
    main()
```
