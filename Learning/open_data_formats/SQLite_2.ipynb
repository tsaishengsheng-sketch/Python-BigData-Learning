```
"""
檔名：SQLite_2.py
功能：開放平台資料格式練習
學習重點：XML、JSON、CSV 讀取與解析應用
"""
import sqlite3
```

```
# 1. 連結你範例一產生的那個 db
con = sqlite3.connect("myDB.db")
c = con.cursor()
```

```
# 2. 建立範例二的資料表 (加上 IF NOT EXISTS 避免重複跑報錯)
c.execute(
    "CREATE TABLE IF NOT EXISTS groups (group_id int primary key not null, group_name char(50) not null);"
)
c.execute("""CREATE TABLE IF NOT EXISTS students (
    student_id int primary key not null, 
    student_name char(50) not null, 
    group_id int not null, 
    FOREIGN KEY (group_id) REFERENCES groups (group_id));""")
```

```
# 3. 手動新增組別 (用 INSERT OR IGNORE 就不怕重複執行)
c.execute("INSERT OR IGNORE INTO groups VALUES (1, '青色之馬');")
c.execute("INSERT OR IGNORE INTO groups VALUES (2, '夢幻之都');")
c.execute("INSERT OR IGNORE INTO groups VALUES (3, '新不了城');")
```

```
# 4. 資料塞進去 (對應課本輸出)
data = [
    (5, "小希", 3),
    (9, "小盛", 3),
    (10, "小霖", 3),
    (11, "小翔", 3),
    (14, "小哲", 3),
]
c.executemany("INSERT OR IGNORE INTO students VALUES (?, ?, ?)", data)
```

```
# 5. 確認存檔
con.commit()
```

```
# 6. 驗證結果：同時查詢原本的 Employee 和新的學生資料
print("--- 原本範例一的 Employee 資料 ---")
for row in c.execute("SELECT * FROM Employee"):
    print(row)
```

```
print("\n--- 範例二追加的第三組學生名單 ---")
cursor = c.execute("""
    SELECT S.student_id, S.student_name, S.group_id, G.group_name 
    FROM students S, groups G 
    WHERE S.group_id = G.group_id AND S.group_id = 3
""")
for record in cursor:
    print(record)
```

```
con.close()
```
