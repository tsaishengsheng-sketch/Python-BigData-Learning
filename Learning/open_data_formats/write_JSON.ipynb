```python
"""
檔名：write_JSON.py
功能：開放平台資料格式練習
學習重點：XML、JSON、CSV 讀取與解析應用
"""
import json
```

```python
# 建立資料
data = {}
# 欄位people 為1串列
data["people"] = []
data["people"].append(
    {
        "id": "1",
        "name": "Peter",
        "country": "Taiwan",
    }
)
data["people"].append(
    {
        "id": "2",
        "name": "Jack",
        "country": "USA",
    }
)
data["people"].append(
    {
        "id": "3",
        "name": "Cindy",
        "country": "Japen",
    }
)
```

```python
# 將資料寫入檔案
with open("JSON_data.json", "w") as outfile:
    json.dump(data, outfile)
```
