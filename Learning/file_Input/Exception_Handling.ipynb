```
"""
檔名：Exception_Handling.py
功能：基礎 Python 練習
學習重點：語法熟悉與邏輯實作
"""
def main():
    try:
        # 1. 嘗試執行這段程式碼
        # eval 可以同時接收多個輸入並自動轉型
        n1, n2 = eval(input('Enter two numbers, seaparated by a comma:'))
        ans = n1 / n2
        print(f"{n1}/{n2}= {ans:.2f}")
        # 如果這裡沒出錯，會直接跳到 finally
        
    except ZeroDivisionError:
        # 2. 捕獲除以 0 的錯誤
        # 當使用者在 n2 輸入 0 時，會執行這裡
        print('denominator can\'t be zero')
        
    except SyntaxError:
        # 3. 捕獲輸入語法錯誤
        # 如果使用者輸入時忘了打逗號（如: 10 2），eval 解析失敗會執行這裡
        print('A comma may be missing in the input')
        
    except:
        # 4. 捕獲「其他所有」未知的錯誤
        # 這是一個保險機制，處理沒想到的意外
        print('Something else went wrong') # 注意：原稿 print('No exception') 邏輯較怪，通常是指發生了未知例外
        
    finally:
        # 5. 不管有沒有出錯，最後「絕對會執行」的一段
        # 通常用來關閉檔案或清理資源
        print('the finally clause is executed')
```

```
main()
```
