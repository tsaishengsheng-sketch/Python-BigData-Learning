```
"""
檔名：Object_Pickle.py
功能：基礎 Python 練習
學習重點：語法熟悉與邏輯實作
"""
import pickle
def main():
    # 1. 以「二進位寫入模式」('wb') 開啟檔案
    # pickle 存的是二進位資料，不是純文字，所以必須加 'b'
    outfile = open('List.dat', 'wb')
    
    # 2. 產生一個 1 到 100 的串列，並使用 pickle.dump() 直接「醃製」存入檔案
    # [x for x in range(1, 101)] 是串列生成式
    pickle.dump([x for x in range(1, 101)], outfile)
    
    # 3. 關閉檔案
    outfile.close()

    # 4. 以「二進位讀取模式」('rb') 開啟檔案
    infile = open('List.dat', 'rb')
    
    # 5. 使用 pickle.load() 將資料「還原」回原本的 Python 物件
    # 注意：讀回來的直接就是 List 型態，不需要 eval() 或 int() 轉換
    data = pickle.load(infile)
    print("--- 從二進位檔案還原的物件 ---")
    print(data)
    
    # 6. 關閉檔案
    infile.close() 
```

```
if __name__ == "__main__":
    main()
```
