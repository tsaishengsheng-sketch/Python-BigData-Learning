```
"""
檔名：File_Score_Statistical_Analyzer.py
功能：基礎 Python 練習
學習重點：語法熟悉與邏輯實作
"""
def main():
    # 1. 取得使用者輸入的檔名並移除前後空白
    f1 = input('Enter a filename: ').strip()

    # 2. 以讀取模式 ('r') 開啟檔案
    infile = open(f1, 'r')

    # 3. 讀取檔案內的所有文字內容
    s = infile.read()

    # 4. 數據解析：
    # s.split()：將字串按空格或換行切開成清單
    # [eval(x) for x in ...]：將切開後的每個字串元素轉為數值型態
    scores = [eval(x) for x in s.split()]
    print("解析出的分數清單：", scores)

    # 5. 進行統計運算並輸出
    # len(scores)：計算總個數
    print(f"There are {len(scores)} scores")
    
    # sum(scores)：計算清單內所有數值的總和
    print(f"The total is {sum(scores)}")
    
    # 計算平均值並格式化為小數點後兩位 (:.2f)
    print(f"The average is {sum(scores)/len(scores):.2f}")

    # 6. 關閉檔案
    infile.close()
```

```
if __name__ == "__main__":
    main()
```
