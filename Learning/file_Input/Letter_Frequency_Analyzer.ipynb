```
"""
檔名：Letter_Frequency_Analyzer.py
功能：基礎 Python 練習
學習重點：語法熟悉與邏輯實作
"""
import os.path
```

```
def main():
    # 1. 取得使用者輸入並移除多餘空格，開啟檔案
    filename = input('Enter a filename: ').strip()
    infile = open(filename, 'r')
    
    # 2. 檢查檔案是否存在
    if os.path.isfile(filename):
        # 建立一個長度 26 的清單 [0, 0, ..., 0]，分別代表 a-z 的出現次數
        counts = 26 * [0]
        
        # 3. 逐行讀取檔案，並統一轉成小寫處理
        for line in infile:
            countLetters(line.lower(), counts)
        
        # 4. 輸出統計結果
        for i in range(len(counts)):
            if counts[i] != 0:
                # chr(ord('a') + i): 將索引還原回字母 (例如 0 變 'a')
                # 最後一段是三元運算子，處理單複數 (time vs times)
                print(chr(ord('a') + i) + ' appears ' + str(counts[i]) + 
                      (' time' if counts[i] == 1 else ' times'))
        
        infile.close()
```

```
def countLetters(line, counts):
    """核心運算：計算字串中每個字母出現的次數"""
    for ch in line:
        if ch.isalpha():  # 只處理英文字母
            # ord(ch) - ord('a'): 利用 ASCII 碼差值計算索引
            # 例如 'a' 是 97, 'b' 是 98 -> 'b' - 'a' = 1 (即 counts[1])
            counts[ord(ch) - ord('a')] += 1
```

```
main()
```
