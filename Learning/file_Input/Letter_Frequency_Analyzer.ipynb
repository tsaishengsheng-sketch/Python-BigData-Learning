```
"""
檔名：Letter_Frequency_Analyzer.py
功能：基礎 Python 練習
學習重點：語法熟悉與邏輯實作
"""
import os.path
```

```
def main():
    # 1. 錯誤處理循環：確保檔案成功開啟
    while True:
        try:
            filename = input('Enter a filename: ').strip()
            infile = open(filename, 'r')
            break # 成功開啟檔案才跳出迴圈
        except IOError:
            # 這裡使用課本格式化字串印出錯誤
            print('File: %s does not exist. Try again' % (filename))

    # --- 注意：以下程式碼必須在 while 迴圈外，否則會邏輯混亂 ---
    
    # 2. 初始化計數器
    counts = 26 * [0]
    
    # 3. 讀取內容並統計
    for line in infile:
        countLetters(line.lower(), counts)
    
    # 4. 輸出統計結果
    for i in range(len(counts)):
        if counts[i] != 0:
            # 依照課本範例，字母與出現次數之間精確顯示
            print(chr(ord('a') + i) + ' appears ' + str(counts[i]) + 
                  (' time' if counts[i] == 1 else ' times'))
    
    infile.close()
```

```
def countLetters(line, counts):
    """核心運算：計算字串中每個字母出現的次數"""
    for ch in line:
        if ch.isalpha():  # 只處理英文字母
            counts[ord(ch) - ord('a')] += 1
```

```
if __name__ == "__main__":
    main()
```
