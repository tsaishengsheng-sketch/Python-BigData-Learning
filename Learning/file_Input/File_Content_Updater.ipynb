```
"""
檔名：File_Content_Updater.py
功能：基礎 Python 練習
學習重點：語法熟悉與邏輯實作
"""
def main():
    # 1. 取得使用者輸入的檔名並移除前後空格
    f1 = input('Enter a filename: ').strip()

    # 2. 以讀取模式 ('r') 開啟目標檔案
    infile = open(f1, 'r')

    # 3. 將檔案內容一次性全部讀入記憶體變數 s 中
    s = infile.read()
    print("--- 讀取到的原始內容 ---")
    print(s)

    # 4. 取得「要被替換」的舊字串與「準備換上」的新字串
    replacedString = input('Enter a string to be replaced: ').strip()
    newContent = input('Enter a new String: ').strip()

    # 5. 使用 replace(舊, 新) 產生替換後的完整字串
    finalString = s.replace(replacedString, newContent)
    print("--- 替換後的預覽內容 ---")
    print(finalString)

    # 6. 關閉讀取流（這是一個好習慣，釋放系統資源）
    infile.close() 

    # 7. 以寫入模式 ('w') 開啟同一個檔案
    # 注意：這會清除檔案原本的所有內容，準備寫入新版內容
    outfile = open(f1, 'w')                   

    # 8. 將取代完成的最終結果 (finalString) 寫回檔案
    # 注意：你原本的程式碼寫成 outfile.write(newString)，那只會存入「新字串」片段
    # 應該要寫入替換後的「全體內容」才會正確
    outfile.write(finalString)
    
    # 9. 關閉輸出流，確保資料完整寫入磁碟
    outfile.close() 
```

```
if __name__ == "__main__":
    main()
```
