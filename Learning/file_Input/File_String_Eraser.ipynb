```
"""
檔名：File_String_Eraser.py
功能：基礎 Python 練習
學習重點：語法熟悉與邏輯實作
"""
def main():
    # 1. 取得使用者輸入的檔名並移除前後空格
    f1 = input('Enter a filename: ').strip()

    # 2. 以讀取模式 ('r') 開啟檔案
    infile = open(f1, 'r')

    # 3. 讀取檔案中的所有內容存入變數 s
    s = infile.read() 
    print("--- 原始檔案內容 ---")
    print(s)

    # 4. 取得使用者想要從檔案中「刪除」的特定字串
    deletedString = input('Enter a string to be deleted: ').strip()

    # 5. 使用 replace 將指定的字串取代為「空字串 ''」，達到刪除效果
    newString = s.replace(deletedString, '')
    print("--- 處理後的內容 ---")
    print(newString)

    # 6. 關閉讀取流
    infile.close() 
    
    # 7. 以寫入模式 ('w') 重新開啟同一個檔案
    # 注意：'w' 模式會直接清空原檔案內容
    outfile = open(f1, 'w')

    # 8. 將處理完的新字串寫入檔案，完成更新
    outfile.write(newString)
    
    # 9. 關閉寫入流，確保資料存入硬碟
    outfile.close() 
```

```
if __name__ == "__main__":
    main()
```
