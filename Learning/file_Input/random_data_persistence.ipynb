```
"""
檔名：random_data_persistence.py
功能：基礎 Python 練習
學習重點：語法熟悉與邏輯實作
"""
import random
```

```
def main():
    lottos = []
    # 1. 開啟檔案準備寫入 ('w')
    outfile = open('LottoNumbers.dat', 'w')
    
    # 產生 6 個不重複的樂透號碼
    for i in range(6):
        num = random.randint(1, 49)
        if num not in lottos:
            lottos.append(num)
            # 重點：檔案 write 只能存字串，所以必須用 str() 轉型
            # 後面加 ' ' 空格是為了讀取時方便區分數字
            outfile.write(str(num) + ' ')
    
    outfile.close() # 關閉檔案存入硬碟
    print("--- 樂透號碼已寫入檔案 ---")

    # 2. 從檔案讀取數據 ('r')
    infile = open('LottoNumbers.dat', 'r')
    
    # read() 會把整份檔案讀成一個大字串，例如 "15 24 28 44 31 20 "
    s = infile.read()
    
    # split() 預設會以「空格」切開字串，回傳一個清單 ['15', '24'...]
    # eval(x) 或 int(x) 將切開後的字串還原回整數型態
    lottoNums = [eval(x) for x in s.split()]
    
    print("從檔案讀取並還原的數字：")
    for num in lottoNums:
        print(num, end = ' ')
        
    infile.close()
```

```
if __name__ == "__main__":
    main()
```
