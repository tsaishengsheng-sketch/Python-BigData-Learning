```
"""
檔名：File_IO_Lab_Basics.py
功能：基礎 Python 練習
學習重點：語法熟悉與邏輯實作
"""
#write .dat 建立檔案(會蓋掉舊檔)
def main():
    outfile = open('fruits.dat', 'w') #寫入
    outfile.write('apple\n')#換行
    outfile.write('Orange\n')
    outfile.write('Banana\n')
    outfile.close()#關閉 存入硬碟
```

```
#read 讀取練習
def read():
    infile = open('fruits.dat','r')
    print('Using read():\n', infile.read())#一次讀完
    infile.close()

    infile = open('fruits.dat', 'r')
    data = infile.read(12) #只讀前12個字元
    print('Using read(12):', repr(data)) # repr 讓 \n 現形
    infile.close()
```

```
#readlines 讀取成清單
def readlines():
    infile = open('fruits.dat', 'r')
    print('Using readlist():', infile.readlines()) # 回傳 ['apple\n', 'Orange\n'...]
    infile.close()
```

```
#readallline 全部讀取(兩種迴圈)
def readAllline():
    infile = open('fruits.dat', 'r')
    line = infile.readline()
    while line != '':                  # 只要還讀得到東西就繼續
        print('While loop:', repr(line))
        line = infile.readline()
    infile.close()

    infile = open('fruits.dat', 'r')
    for line in infile:         
        print('For loop:', repr(line))
    infile.close()
```

```
#isfile() & append
import os.path
def os_read():
    infile = open('fruits.dat', 'r')
    if os.path.isfile('fruits.dat'):
        print('Using read(): \n', infile.read())
    infile.close()
```

```
def append():
    # 先檢查檔案在不在，再追加
    if os.path.isfile('fruits.dat'):
        outfile = open('fruits.dat', 'a') # 'a' 代表追加模式
        outfile.write('Kiwi\n')           # 在原本的 Banana 後面加 Kiwi
        outfile.close()

    # 讀取最終結果
    with open('fruits.dat', 'r') as infile: 
        print('Final Results:\n', infile.read())
```

```
main()
read()
readlines()
readAllline()
append()
```
