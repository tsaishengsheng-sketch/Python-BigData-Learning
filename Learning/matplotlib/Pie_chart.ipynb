```
"""
檔名：Pie_chart.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
import matplotlib.pyplot as plt
import numpy as np
import random
```

```
rdata = []
count = [0] * 6
```

```
for i in range(1, 51):
    n = random.randint(1, 5)
    rdata.append(n)
```

```
for i in range(len(rdata)):
    if rdata[i] == 1:
        count[1] += 1
    elif rdata[i] == 2:
        count[2] += 1
    elif rdata[i] == 3:
        count[3] += 1
    elif rdata[i] == 4:
        count[4] += 1
    else:
        count[5] += 1
```

```
count.sort()
print(count)
```

```
percentage = "%.1f%%"
numbers = [0, 1, 2, 3, 4, 5]
labels = [str(num) for num in numbers]
colors = ["w", "r", "g", "m", "b", "y"]
```

```
plt.pie(count, labels=labels, autopct=percentage, colors=colors)
plt.title("Pie chart")
plt.ylabel("")
plt.show()
```
