```
"""
檔名：line_chart.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
import matplotlib.pyplot as plt
import numpy as np
```

```
# 20個亂數
rdata = np.random.randn(20)
```

```
plt.xticks(np.arange(1, 20, 5))
plt.title("My first matplotlib program.")
plt.plot(rdata, "bD--", linewidth=3)
```
