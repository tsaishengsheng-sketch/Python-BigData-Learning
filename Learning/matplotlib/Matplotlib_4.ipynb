```
"""
檔名：Matplotlib_4.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
import numpy as np
import matplotlib.pyplot as plt
```

```
# 隨機產生50個 0~10 的浮點數
x = np.random.rand(50) * 10
```

```
# 隨機產生50個 1~100的整數
y = np.random.randint(1, 101, 50)
```

畫出散佈 點大小為80 點樣式:上三角形
顏色隨機介於0~1之間，透明度0.7

```
plt.scatter(x, y, s=80, marker="^", c=np.random.rand(50), alpha=0.7)
```

```
# x軸名稱
plt.xlabel("x_value", size=14)
# y軸名稱
plt.ylabel("y_value", size=14)
# x軸刻度 0~10
plt.xticks(np.arange(11))
# y軸刻度 0~100 間隔20
plt.yticks(np.arange(0, 101, 20))
# 設定圖表標題
plt.title("Scatter chart", size=20)
```

```
# 顯示圖
plt.show()
```
