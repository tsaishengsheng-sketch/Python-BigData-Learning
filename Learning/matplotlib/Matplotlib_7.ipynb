```
"""
檔名：Matplotlib_7.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
import matplotlib.pyplot as plt
```

```
# A 公司各部門及員工數量
companyA = {
    "Personnel Department": 8,
    "Accounting Department": 5,
    "Design Department": 10,
    "Marketing Department": 12,
    "Purchasing Department": 5,
}
```

```
# B 公司各部門及其員工數量
companyB = {
    "Personnel Department": 10,
    "Accounting Department": 8,
    "Design Department": 20,
    "Marketing Department": 18,
    "Purchasing Department": 12,  # 修正：移除多餘空格
}
```

```
companyA_values = list(companyA.values())
companyB_values = list(companyB.values())
keys = companyA.keys()
colors = ["violet", "yellow", "skyblue", "Lightcoral", "Lightgreen"]
```

```
# 修正：將變數名稱統一為全小寫 explode
explode = [0, 0, 0, 0.1, 0]
autopct = "%.1f%%"
startangle = 90
```

```
# 繪出A公司的圓形圖
fig1, ax1 = plt.subplots()
ax1.pie(
    companyA_values,
    labels=keys,
    explode=explode,  # 這裡現在能正確對應了
    colors=colors,
    autopct=autopct,
    startangle=startangle,
)
ax1.axis("equal")
ax1.set_title("Company A", size=20)
```

```
# 繪出B公司的圓形圖
fig2, ax2 = plt.subplots()
ax2.pie(
    companyB_values,
    labels=keys,
    explode=explode,
    colors=colors,
    autopct=autopct,
    startangle=startangle,
)
ax2.axis("equal")
ax2.set_title("Company B", size=20)
```

```
plt.show()
```
