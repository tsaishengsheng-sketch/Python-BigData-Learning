```
"""
檔名：q2_bar_chart.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
import matplotlib.pyplot as plt
import pydataset
import pandas as pd
```

```
# 使用 pydataset 的 iris 資料集
df = pd.DataFrame(pydataset.data("iris"))
```

```
# 計算每個品種的平均花瓣長度
species_avg = df.groupby("Species")["Petal.Length"].mean()
plt.rcParams["font.family"] = "Arial Unicode MS"
plt.title("各品種花瓣平均長度")
plt.xlabel("品種")
plt.ylabel("平均花瓣長度")
plt.bar(
    species_avg.index, species_avg.values, color=["skyblue", "lightgreen", "lightcoral"]
)
plt.show()
```
