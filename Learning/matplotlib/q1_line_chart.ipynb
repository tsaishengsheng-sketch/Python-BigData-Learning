```
"""
檔名：q1_line_chart.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
import matplotlib.pyplot as plt
import numpy as np
```

```
np.random.seed(42)
```

```
# 產生兩組簡單資料
x = np.arange(1, 11)
y1 = np.random.randint(10, 50, 10)
y2 = np.random.randint(10, 50, 10)
```

```
plt.title("折線圖範例")
plt.xlabel("X 軸")
plt.ylabel("Y 軸")
plt.plot(x, y1, "-o", label="資料1", color="blue")
plt.plot(x, y2, "-o", label="資料2", color="red")
plt.legend()
plt.show()
```
