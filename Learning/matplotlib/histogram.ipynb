```
"""
檔名：histogram.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
import matplotlib.pyplot as plt
import numpy as np
import random
```

```
x = np.arange(1, 11, 1)
rdata = []
for i in range(50):
    n = random.randint(1, 10)
    rdata.append(n)
```

```
rdata.sort()
print(rdata)
plt.hist(rdata)
plt.title("Histogram")
plt.xticks(x)
```

```
plt.show()
```
