```
"""
檔名：q4_scatter.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
import matplotlib.pyplot as plt
import pydataset
import pandas as pd
import numpy as np
```

```
np.random.seed(42)
```

```
# 使用 pydataset 的 cars 資料集
df = pd.DataFrame(pydataset.data("cars"))
```

```
# 隨機產生每個點的顏色
colors = np.random.rand(len(df), 3)
```

```
plt.rcParams["font.family"] = "Arial Unicode MS"
plt.title("車速與剎車距離散佈圖")
plt.xlabel("速度 (speed)")
plt.ylabel("剎車距離 (dist)")
plt.scatter(df["speed"], df["dist"], color=colors, alpha=0.5)
plt.show()
```
