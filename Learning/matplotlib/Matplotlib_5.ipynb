```
"""
檔名：Matplotlib_5.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
import numpy as np
import random
from matplotlib import pyplot as plt
```

```
np.random.seed(100)
# 學生分數
data = [np.random.randint(0, 100) for _ in range(100)]
```

```
# 將分數串列轉換為np陣列
data = np.array(data)
```

```
# 等差級數決定方盒的數量 為20
bins = np.linspace(np.ceil(min(data)), np.floor(max(data)), 20)
```

```
# 繪出直方圖，顏色為橘色，透明度為0.5
plt.hist(data, bins, color="orange", alpha=0.5)
```

```
# x軸刻度
plt.xlim([0, 100])
```

```
# x軸名稱
plt.xlabel("Score", size=14)
```

```
# y軸名稱
plt.ylabel("Quantity", size=14)
```

```
# 設定圖標題
plt.title("Stduent Scores Distribution", size=20)
```

```
# 顯示圖
plt.show()
```
