```
"""
檔名：q3_histogram.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```
np.random.seed(42)
```

```
# 產生 1000 組標準常態分配隨機變數（平均值0，標準差1）
data = np.random.normal(0, 1, 1000)
```

```
plt.rcParams["font.family"] = "Arial Unicode MS"
plt.title("標準常態分配直方圖")
plt.xlabel("數值")
plt.ylabel("頻率")
plt.hist(data, bins=30, color="skyblue", edgecolor="black")
plt.show()
```
