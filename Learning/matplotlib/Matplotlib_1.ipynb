```
"""
檔名：Matplotlib_1.py
功能：資料視覺化練習
學習重點：折線圖、長條圖、圓餅圖與圖表樣式設定
"""
from matplotlib import pyplot as plt
import numpy as np
import random
```

```
# 學生分數
scores = [random.randint(0, 100) for _ in range(70)]
```

range_count [0]: range0~19
range_count [1]: range20~39
range_count [0]: range40~59
range_count [0]: range60~79
range_count [0]: range80~100

```
# 以0初始化計數串列
range_count = [0] * 5
```

```
# 計數過程
for score in scores:
    if score < 20:
        range_count[0] += 1
    elif score < 40:
        range_count[1] += 1
    elif score < 60:
        range_count[2] += 1
    elif score < 80:
        range_count[3] += 1
    else:
        range_count[4] += 1
```

```
# Y軸標籤
index = np.arange(0, 25, 5)
```

```
# X軸標籤
Labels = ["0~19", "20~39", "40~59", "60~79", "80~100"]
```

```
# 畫出長條圖
plt.bar(index, range_count, width=2)
```

```
# 設定X軸名稱
plt.xlabel("Range", fontsize=14)
```

```
# 設定Y軸名稱
plt.ylabel("Range", fontsize=14)
```

```
# 設定x軸標籤
plt.xticks(index, Labels)
```

```
# 設定y軸標籤
plt.yticks(index)
```

```
# 設定圖名稱
plt.title("Scoer ranges count", fontsize=20)
```

```
# 顯示圖
plt.show()
```
